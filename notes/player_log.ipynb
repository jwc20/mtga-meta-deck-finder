```python
client_messages = game_object.get('greToClientEvent').get('greToClientMessages')
```

```python
results = []
for i, message in enumerate(client_messages):
    game_state_message = message.get('gameStateMessage')
    if not isinstance(game_state_message, dict):
        continue
    
    game_objects = game_state_message.get('gameObjects')
    actions = game_state_message.get('actions')
    
    results.append((i, game_objects, actions))
    
    # if isinstance(game_objects, list) and isinstance(actions, list):
    #     results.append((i, game_objects, actions))
    
pprint(results)
print(f"Found {len(results)} game states with both objects and actions")
```

```python
def map_actions_to_cards(game_objects: list, actions: list) -> list:
    instance_to_card = {obj['instanceId']: obj for obj in game_objects}
    
    result = []
    for action_entry in actions:
        action = action_entry['action']
        instance_id = action.get('instanceId')
        card = instance_to_card.get(instance_id)
        
        result.append({
            'seat_id': action_entry['seatId'],
            'action_type': action['actionType'],
            'instance_id': instance_id,
            'ability_id': action.get('abilityGrpId'),
            'mana_cost': action.get('manaCost'),
            'card_grp_id': card['grpId'] if card else None,
            'card_types': card.get('cardTypes') if card else None,
        })
    
    return result
```

```python
for i, game_objects, actions in results[:1]:
    actions = map_actions_to_cards(game_objects, actions)
    print(i)
    pprint(actions)
    print("########################")
    print(len(actions))
```
